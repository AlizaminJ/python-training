#### Adding a new row is a bit complicated, you have to transpose first and transpose back

```python
df7.set_index("Address", inplace = True) 
df7
```

```python
df7_t = df7.T
df7_t
```

```python
df7_t.shape
```

```python
df7_t["My Address"] = ["My City", "My Country", 10, 7, "My Shop", "My State", "My Continent"]
df7_t
```

```python
df7 = df7_t.T
```

```python
df7
```

### Shorter way to do that

### <a href="https://thispointer.com/python-pandas-how-to-add-rows-in-a-dataframe-using-dataframe-append-loc-iloc/"> Useful article </a>

```python
df8=df7.append({"City": "Your City", "Country":"Your Country"}, ignore_index=True)
df8
```

```python
# List of series  
listOfSeries = [pandas.Series(['Baku', 'Azerbaijan', 9, 2, "Baku", "Baku", "Asia"], index=df7.columns ) ,
                pandas.Series(['Istanbul', 'Turkey', 10, 3, "Istanbul", "Istanbul", "Asia"], index=df7.columns ) ,
                pandas.Series(['Tbilisi', 'Georgia', 11, 4, "Tbilisi", "Tbilisi", "Asia"], index=df7.columns ) ]
```

```python
df9 = df7.append(listOfSeries , ignore_index=True)
```
